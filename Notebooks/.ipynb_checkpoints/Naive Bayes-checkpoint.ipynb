# 1. Load Data

The first thing we need to do is load our data file. The data is in CSV format with a header line. The data has been uploaded to my github repository from where I am reading the data from.

```python
bot_data_url = 'https://raw.githubusercontent.com/bijilbabu/Am-I-A-Twitter-Bot/master/Training%20Data/final/bots_data.csv'
non_bot_data_url = 'https://raw.githubusercontent.com/bijilbabu/Am-I-A-Twitter-Bot/master/Training%20Data/final/nonbots_data.csv'

#Read the csv
bots_data = pd.read_csv(bot_data_url)
non_bots_data = pd.read_csv(non_bot_data_url)

# Merge the different sets , then randomize the dataset, then reset index to rebuild the index.
dataset = pd.concat([bots_data, non_bots_data]).sample(frac=1).reset_index(drop=True)

print "Shape of the dataset %s" % str(dataset.shape)

# bot distribution
print dataset.groupby('bot').size()
```

# 2. Finding relationship with among the data.

```python
# keeping only the features being used

features_dataset = dataset[[
#     'screen_name', 
                        'followers_count', 
                        'friends_count', 
                        'listedcount', 
                        'favourites_count', 
                        'verified', 
                        'statuses_count', 
#                         'lang', 
                        'default_profile', 
                        'default_profile_image', 
#                         'has_extended_profile',
                        'bot']]
```

```python
%matplotlib inline
colors=['blue','green']
scatter_matrix(features_dataset, figsize=[40,40],marker='x', c=features_dataset.bot.apply(lambda x:colors[x]))
```

Next we need to split the data into a training dataset that Naive Bayes can use to make predictions and a test dataset that we can use to evaluate the accuracy of the model. We need to split the data set randomly into train and datasets with a ratio of 67% train and 33% test (this is a common ratio for testing an algorithm on a dataset).

```python
dataset_array = features_dataset.values
features = dataset_array[:,0:8]
classification = map(str, dataset_array[:,8])

validation_size = 0.20
seed = 7
scoring = 'accuracy'
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(features, classification, test_size=validation_size, random_state=seed)

```

```python
def plot_confusion_matrix(cm, classes, normalize=False,title='Confusion matrix',cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')
    
    print(cm)
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")
        
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
## Gaussian Naive bayes

nb = GaussianNB()
nb.fit(X_train, Y_train)
predictions = nb.predict(X_validation)
print(accuracy_score(Y_validation, predictions))
plot_confusion_matrix(confusion_matrix(Y_validation, predictions), classes=['bot', 'non-bot'], title='Confusion matrix, without normalization')
print(classification_report(Y_validation, predictions))
```

```python

# Random Forest
rf = RandomForestClassifier().fit(X_train, Y_train)
rf_predict = rf.predict(X_validation)

print(accuracy_score(Y_validation, rf_predict))
plot_confusion_matrix(confusion_matrix(Y_validation, rf_predict), classes=['bot', 'non-bot'], title='Confusion matrix, without normalization')
print(classification_report(Y_validation, rf_predict))
```

```python
#Bernoulii Naive Bayes
bnb = BernoulliNB().fit(X_train, Y_train)

bnbPredictions = bnb.predict(X_validation)

print(accuracy_score(Y_validation, bnbPredictions))
plot_confusion_matrix(confusion_matrix(Y_validation, bnbPredictions), classes=['bot', 'non-bot'], title='Confusion matrix, without normalization')
print(classification_report(Y_validation, bnbPredictions))
```
